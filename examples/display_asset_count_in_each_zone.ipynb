```python
# import dependencies
from rdtl.tools import display_asset_count_in_each_zone
import pandas as pd
from datetime import datetime

# read in the pixel data
df = pd.read_excel('tote_shipment.xlsx', parse_dates = True)
df = df.rename(str.lower,axis='columns')

# get current time
current_datetime = pd.to_datetime(datetime.now())
```

```python
# display the asset count in all zones
display_asset_count_in_each_zone(pixel_dataframe=df, current_time_est=current_datetime)
```

```python
# get a different time
date_string = '8/2/2022 12:24:05 PM'
date_format = '%m/%d/%Y %I:%M:%S %p'
datetime_object = datetime.strptime(date_string, date_format)

# display the asset count in all zones
display_asset_count_in_each_zone(pixel_dataframe=df, current_time_est=datetime_object)
```

```python
# display the asset count in the zones within the 'Store' site
display_asset_count_in_each_zone(pixel_dataframe=df, current_time_est=current_datetime, site='Store')
```
